```python
from operator import itemgetter

PEOPLE = [('Donald', 'Trump', 7.85),
          ('Vladimir', 'Putin', 3.626),
          ('Jinping', 'Xi', 10.603)]


def format_sort_records(list_of_records):
    formated_records = []
    # {position:nombre de characteres}
    template = '{1:10} {0:10} {2:10.2f}'
    # trie d’abord par le nom de famille, et si deux personnes ont le même nom de famille, utilise le prénom pour départager
    for line in sorted(list_of_records, key=itemgetter(1, 0)):
        formated_records.append(template.format(*line))
    return formated_records


```

```python
format_sort_records(PEOPLE)
```

```python
# Reimplement this exercise using namedtuple objects defined in the collections module.

from collections import namedtuple

Person = namedtuple("Person", ["first", "last", "score"])

PEOPLE = [
    Person("Donald", "Trump", 7.85),
    Person("Vladimir", "Putin", 3.626),
    Person("Jinping", "Xi", 10.603),
]

def format_sort_records_namedTuple(records):
    formated_records = []
    # on trie par last (index 1), puis par first (index 0)
    for line in sorted(records, key=lambda x: (x.last, x.first)):
        formated_records.append(f"{line.last:10} {line.first:10} {line.score:10.2f}")
    return formated_records
```

```python
format_sort_records_namedTuple(PEOPLE)
```
